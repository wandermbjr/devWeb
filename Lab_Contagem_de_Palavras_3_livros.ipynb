```
import requests
from bs4 import BeautifulSoup
import nltk
```

```
r1 = requests.get("https://www.gutenberg.org/files/83/83-h/83-h.htm")     # From the Earth to the Moon, by Jules Verne
r1.encoding = 'utf-8'

html1 = r1.text

#print(html1[0:2000])

r2 = requests.get("https://www.gutenberg.org/files/155/155-h/155-h.htm")   # The Moonstone, by Wilkie Collins
r2.encoding = 'utf-8'

html2 = r2.text

#print(html2[0:2000])


r3 = requests.get("https://www.gutenberg.org/files/69102/69102-h/69102-h.htm")   # The Moon, by Oliver Cummings Farrington
r3.encoding = 'utf-8'

html3 = r3.text

#print(html2[0:2000])









```

```
soup = BeautifulSoup(html1, 'html.parser')

text1 = soup.get_text()

#print(text1[32000:34000])


soup = BeautifulSoup(html2, 'html.parser')

text2 = soup.get_text()

#print(text2[32000:34000])

soup = BeautifulSoup(html3, 'html.parser')

text3 = soup.get_text()

#print(text2[32000:34000])
```

```
tokenizer = nltk.tokenize.RegexpTokenizer('\w+')

tokens_1 = tokenizer.tokenize(text1)

#print(tokens_1[0:14])


tokens_2 = tokenizer.tokenize(text2)

#print(tokens_2[0:14])

tokens_3 = tokenizer.tokenize(text3)

#print(tokens_3[0:14])
```

```
words_1 = []

for word in tokens_1:
  words_1.append(word.lower())

#print(words_1[0:8])

words_2 = []

for word in tokens_2:
  words_2.append(word.lower())

#print(words_1[0:14])

words_3 = []

for word in tokens_3:
  words_3.append(word.lower())

#print(words_3[0:14])
```

```
from nltk.corpus import stopwords
nltk.download('stopwords')

sw_1 = stopwords.words('english')

sw_2 = stopwords.words('english')

sw_3 = stopwords.words('english')






```

```
words_ns_1 =  [word for word in words_1 if word not in sw_1]

words_ns_2 =  [word for word in words_2 if word not in sw_2]

words_ns_3 =  [word for word in words_3 if word not in sw_3]
```

```
#print(words_ns_1[0:8])

#print(words_ns_2[0:8])

#print(words_ns_3[0:8])
```

```
freqdist_1 = nltk.FreqDist(words_ns_1)

freqdist_2 = nltk.FreqDist(words_ns_2)

freqdist_3 = nltk.FreqDist(words_ns_3)
```

```
freqdist_1

freqdist_2

freqdist_3
```

```
%matplotlib inline

freqdist_1.plot(500)

freqdist_2.plot(500)

freqdist_3.plot(500)
```

```
# Como podemos verificar nos 3 graficos , todos estão respeitando a lei de zipf.
# As palavras mais frequentes, em cada uma das analises estão relacionadas aos enredos de cada livro respectivamente.

```
